```python
import onnxruntime as ort
import numpy as np

# Load the ONNX model file
model_path = "emotion_model.onnx"
session = ort.InferenceSession(model_path, providers=['CPUExecutionProvider'])

print(session.get_inputs()[0].name)
print(session.get_inputs()[0].shape)
print(session.get_inputs()[0].type)
```

```python

for inp in session.get_inputs():
    print(inp.name, inp.shape, inp.type)
```

```python


# Prepare input data as numpy arrays (example shapes and dtype)

input_name = 'unk__982'
input_data = np.random.rand(1, 48, 48, 1).astype(np.float32)
# Run inference

input_name = session.get_inputs()[0].name  # Should now be 'input'
outputs = session.run(None, {input_name: input_data})

print(outputs)

```

```python
pred = outputs[0]
predicted_class = pred.argmax(axis=1)[0]
print("Predicted class index:", predicted_class)
```

```python
categories = ["angry", "disgust", "fear", "happy", "sad", "surprise", "neutral"]
print("Predicted emotion:", categories[predicted_class])
```
